#### XGBoost Classifier

```python
import time
import numpy as np
import xgboost as xgb
from sklearn.model_selection import GridSearchCV

# ------------------------------------------------------------
# 1) Calcul du poids de la classe positive (churn=1) : utile si dataset déséquilibré
#    scale_pos_weight = (nb_neg / nb_pos)
# ------------------------------------------------------------
y_train_np = np.asarray(train_labels)
neg = np.sum(y_train_np == 0)
pos = np.sum(y_train_np == 1)

scale_pos_weight = (neg / pos) if pos > 0 else 1.0
print(f"neg={neg}, pos={pos}, scale_pos_weight={scale_pos_weight:.3f}")

# ------------------------------------------------------------
# 2) Grille d'hyperparamètres (raisonnable et efficace)
# ------------------------------------------------------------
param_grid_xgb = {
    "n_estimators": [200, 300, 500],
    "max_depth": [3, 4, 5, 6, 7],
    "learning_rate": [0.03, 0.05, 0.1],
    "min_child_weight": [1, 3, 5, 7],
    "subsample": [0.8, 1.0],
    "colsample_bytree": [0.8, 1.0],
    # Option pro (tu peux activer si besoin) :
    "gamma": [0, 0.1, 0.3],
    "reg_alpha": [0, 0.1],
    "reg_lambda": [1, 2]
}

# ------------------------------------------------------------
# 3) Modèle XGBoost
# ------------------------------------------------------------
xgb_c = xgb.XGBClassifier(
    random_state=42,
    n_jobs=-1,
    eval_metric="logloss",
    use_label_encoder=False,   # OK même si maintenant déprécié selon versions
    scale_pos_weight=scale_pos_weight,
    tree_method="hist"         # accélère sur CPU (en général)
)

# ------------------------------------------------------------
# 4) GridSearchCV (même logique que ton RF)
# ------------------------------------------------------------
grid_search_xgb = GridSearchCV(
    estimator=xgb_c,
    param_grid=param_grid_xgb,
    scoring="f1",
    cv=5,
    n_jobs=-1,
    verbose=1
)

# ------------------------------------------------------------
# 5) Fit (conversion sparse -> dense si nécessaire)
# ------------------------------------------------------------
start_time = time.time()

X_train_in = train_features.toarray() if hasattr(train_features, "toarray") else train_features
grid_search_xgb.fit(X_train_in, train_labels)

elapsed_time = time.time() - start_time

# ------------------------------------------------------------
# 6) Résultats
# ------------------------------------------------------------
print(f"Temps de calcul: {elapsed_time:.2f} secondes")
print("Meilleurs paramètres:", grid_search_xgb.best_params_)
print("Meilleur score CV (F1):", round(grid_search_xgb.best_score_, 3))
print(grid_search_xgb.best_estimator_.get_params())
```

### Hyperparameter Tuning (Optuna) & Experiment Tracking (MLFlow)

```python
import optuna
import numpy as np
from lightgbm import LGBMClassifier
from sklearn.metrics import recall_score, precision_score, f1_score

# ------------------------------------------------------------
# Hypothèses : variables DÉJÀ existantes
# ------------------------------------------------------------
# train_features
# train_labels
# X_test_in
# y_test

THRESHOLD = 0.5120157

X_train_in = train_features
y_train_np = np.asarray(train_labels)

# ------------------------------------------------------------
# Optuna objective function
# ------------------------------------------------------------
def objective(trial):

    params = {
        "n_estimators": trial.suggest_categorical("n_estimators", [100, 200, 300, 500]),
        "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.2, log=True),
        "max_depth": trial.suggest_int("max_depth", -1, 10),
        "num_leaves": trial.suggest_int("num_leaves", 16, 256),
        "min_child_samples": trial.suggest_int("min_child_samples", 10, 100),
        "subsample": trial.suggest_float("subsample", 0.6, 1.0),
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.6, 1.0),
        "reg_alpha": trial.suggest_float("reg_alpha", 0.0, 5.0),
        "reg_lambda": trial.suggest_float("reg_lambda", 0.0, 5.0),

        # gestion du déséquilibre
        "class_weight": "balanced",

        # reproductibilité / perf
        "random_state": 42,
        "n_jobs": -1
    }

    model = LGBMClassifier(**params)
    model.fit(X_train_in, y_train_np)

    proba = model.predict_proba(X_test_in)[:, 1]
    y_pred = (proba >= THRESHOLD).astype(int)

    # 👉 métrique optimisée (comme ton code XGB)
    return recall_score(y_test, y_pred, pos_label=1)

# ------------------------------------------------------------
# Run Optuna
# ------------------------------------------------------------
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=30)

print("Best params:", study.best_params)
print("Best recall:", study.best_value)
```

```python
import time
import numpy as np
from lightgbm import LGBMClassifier
from sklearn.metrics import classification_report

# ------------------------------------------------------------
# Hypothèses : variables déjà existantes
# ------------------------------------------------------------
# study            -> Optuna study
# X_train_in       -> features train
# y_train_np       -> labels train (numpy)
# X_test_in        -> features test
# y_test           -> labels test
# ------------------------------------------------------------

THRESHOLD = np.float32(0.5120157)

# Récupérer les meilleurs paramètres trouvés par Optuna
best_params = study.best_params.copy()

# Ajouter les paramètres fixes du modèle final
best_params.update({
    "class_weight": "balanced",
    "random_state": 42,
    "n_jobs": -1
})

# Créer le modèle final LightGBM
lgbm = LGBMClassifier(**best_params)

# ------------------------------------------------------------
# Training timer
# ------------------------------------------------------------
start_time = time.time()
lgbm.fit(X_train_in, y_train_np)
elapsed_time = time.time() - start_time
print(f"Training time: {elapsed_time:.2f} seconds")
```

```python
# ------------------------------------------------------------
# Prediction timer
# ------------------------------------------------------------
start_pred_time = time.time()
proba = lgbm.predict_proba(X_test_in)[:, 1]
y_pred = (proba >= THRESHOLD).astype(int)
elapsed_pred_time = time.time() - start_pred_time
print(f"Prediction time: {elapsed_pred_time:.2f} seconds")
```
